```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from pickle import dump

prep_train = pd.read_csv('/content/drive/My Drive/CMCL Shared Task/preprocessed_dataset.csv')

scaler = MinMaxScaler()

keys = train_data.keys()[3:8]

a = train_data.loc[:, keys]                    
prep_train.update(a)                                  # Overwrite standardized target variables with actual variables
b = scaler.fit_transform(prep_train.loc[:, keys])     # Min Max scaling

b = pd.DataFrame(b, columns = keys)            
prep_train.update(b)                                  # Updating values to Min max scaled values
prep_train

dump(scaler, open('/content/drive/My Drive/CMCL Shared Task/MinMaxScaler.pkl', "wb"))
```

```
gb = prep_train.groupby("sentence_id")
gb.get_group(0)
```

```
import random
seed = 10

np.random.seed(seed)
random.seed(seed)

sents = random.sample(range(0, 800), 800)
train_sents = sents[:int(0.8*len(sents))]
val_sents = sents[int(0.8*len(sents)):]
```

```
train = pd.DataFrame()
val = pd.DataFrame()

for sent in train_sents:
  train = pd.concat([train, gb.get_group(sent)], axis = 0)

for sent in val_sents:
  val = pd.concat([val, gb.get_group(sent)], axis = 0)
```

```
train.to_csv("/content/drive/My Drive/CMCL Shared Task/train_split.csv", index = False)
val.to_csv("/content/drive/My Drive/CMCL Shared Task/val_split.csv", index = False)
```

```
roberta_submission = pd.read_csv('/content/drive/My Drive/CMCL Shared Task/MinMax_roberta_21_submission.csv')
lstm_submission = pd.read_csv("/content/drive/My Drive/CMCL Shared Task/minmax_numbatch_BiLSTM_submission.csv")
bert_submission = pd.read_csv("/content/drive/MyDrive/CMCL Shared Task/BertTokenMinMax.csv")
```

```
roberta_submission.to_csv("roberta_submission.txt", index = False)
```

```
roberta_pred = roberta_submission.loc[:, ["nFix", "FFD", "GPT", "TRT", "fixProp"]]
lstm_pred = lstm_submission.loc[:, ["nFix", "FFD", "GPT", "TRT", "fixProp"]]
bert_pred = bert_submission.loc[:, ["nFix", "FFD", "GPT", "TRT", "fixProp"]]
train_targets = train_data.loc[:, ["nFix", "FFD", "GPT", "TRT", "fixProp"]]
```

```
print("Train Means:\n", train_targets.mean(axis=0))
print("Roberta Means:\n", roberta_pred.mean(axis=0))
print("LSTM Means:\n", lstm_pred.mean(axis=0))
print("BERT Means:\n", bert_pred.mean(axis=0))
```

```
keys = ["nFix", "FFD", "GPT", "TRT", "fixProp"]

robert = {}
lstm = {}
bert = {}
total = {}

robert["nFix"] = 0.883
lstm["nFix"] = 0.634
bert["nFix"] = 0.743

robert["FFD"] = 0.913
lstm["FFD"] = 0.720
bert["FFD"] = 0.852

robert["GPT"] = 0.738 
lstm["GPT"] = 0.593
bert["GPT"] = 0.457

robert["TRT"] =0.854
lstm["TRT"] = 0.519
bert["TRT"] = 0.692

robert["fixProp"] = 0.946
lstm["fixProp"] = 0.756
bert["fixProp"] = 0.917

for key in keys:
  total[key] = robert[key]+lstm[key]+bert[key]
```

```
final_submission = pd.DataFrame()
for key in keys:
  final_submission[key] = (robert[key]*roberta_submission[key]+lstm[key]*lstm_submission[key]+bert[key]*bert_submission[key])/total[key]

final_submission
```

```
test_data = pd.read_csv('/content/drive/My Drive/CMCL Shared Task/test_data.csv')
test_data
```

```
submission = pd.concat([test_data, final_submission], axis = 1)
submission
```

```
submission.to_csv('/content/drive/My Drive/CMCL Shared Task/Ensemble_RBL_submission.csv', index = False)
submission.to_csv('answer.txt', index = False)
```
